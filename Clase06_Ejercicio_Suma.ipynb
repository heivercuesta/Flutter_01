####Haciendo uso de Flutter, desarrollarun aplicativo quepermita recibir dos números por pantalla y que al dar CLic en el Boton Ejecutar sume los datos y muestre el resultado en Pantalla

```python
/
import 'package:flutter/material.dart';

void main() {
  runApp(const SumaApp());
}

// -----------------------------------------------------------------
// 1. WIDGET PRINCIPAL Y GESTIÓN DE ESTADO
// -----------------------------------------------------------------

class SumaApp extends StatelessWidget {
  const SumaApp({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'Taller de Suma Flutter',
      theme: ThemeData(
        primarySwatch: Colors.indigo,
        useMaterial3: true,
      ),
      home: const SumaScreen(),
    );
  }
}

// Usamos StatefulWidget para poder actualizar el resultado en la pantalla.
class SumaScreen extends StatefulWidget {
  const SumaScreen({super.key});

  @override
  State<SumaScreen> createState() => _SumaScreenState();
}

class _SumaScreenState extends State<SumaScreen> {
  // Controladores de TextField: Necesarios para leer el texto ingresado.
  final TextEditingController _num1Controller = TextEditingController();
  final TextEditingController _num2Controller = TextEditingController();

  // Variable de Estado: Almacena el resultado que se muestra con el widget Text.
  double _resultado = 0;
  String _error = '';

  // -----------------------------------------------------------------
  // 2. LÓGICA DE SUMA
  // -----------------------------------------------------------------
  void _calcularSuma() {
    // Usamos setState para notificar a Flutter que el estado ha cambiado.
    setState(() {
      _error = ''; // Limpiar errores anteriores

      // Intentar parsear los textos a números (double)
      final double? num1 = double.tryParse(_num1Controller.text);
      final double? num2 = double.tryParse(_num2Controller.text);

      if (num1 != null && num2 != null) {
        // Si ambos son números válidos, realizar la suma
        _resultado = num1 + num2;
      } else {
        // Si hay un error de formato
        _resultado = 0;
        _error = '⚠️ Por favor, ingresa números válidos en ambos campos.';
      }
    });
  }

  // -----------------------------------------------------------------
  // 3. ESTRUCTURA DE LA INTERFAZ DE USUARIO (UI)
  // -----------------------------------------------------------------
  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(
        title: const Text('Calculadora de Suma Simple'),
        backgroundColor: Theme.of(context).primaryColor,
      ),
      // Padding envuelve Column para agregar márgenes al contenido
      body: Padding(
        padding: const EdgeInsets.all(32.0),
        child: Column( // Organiza los widgets verticalmente
          mainAxisAlignment: MainAxisAlignment.center, // Centra el contenido verticalmente
          children: <Widget>[

            // Campo de texto para el Primer Número (TextField)
            TextField(
              controller: _num1Controller,
              keyboardType: TextInputType.number, // Teclado numérico
              decoration: const InputDecoration(
                labelText: 'Primer Número',
                border: OutlineInputBorder(),
              ),
            ),

            const SizedBox(height: 20), // Espacio vertical entre widgets

            // Campo de texto para el Segundo Número (TextField)
            TextField(
              controller: _num2Controller,
              keyboardType: TextInputType.number, // Teclado numérico
              decoration: const InputDecoration(
                labelText: 'Segundo Número',
                border: OutlineInputBorder(),
              ),
            ),

            const SizedBox(height: 30),

            // Botón para ejecutar la suma (ElevatedButton)
            ElevatedButton(
              onPressed: _calcularSuma, // Llama a la función de suma
              style: ElevatedButton.styleFrom(
                padding: const EdgeInsets.symmetric(horizontal: 40, vertical: 15),
              ),
              child: const Text(
                'Sumar Números',
                style: TextStyle(fontSize: 18),
              ),
            ),

            const SizedBox(height: 40),

            // Muestra el resultado (Text)
            const Text(
              'Resultado:',
              style: TextStyle(fontSize: 24, fontWeight: FontWeight.w300),
            ),

            Text(
              _resultado.toStringAsFixed(2), // Muestra el resultado formateado
              style: TextStyle(
                fontSize: 48,
                fontWeight: FontWeight.bold,
                color: Theme.of(context).primaryColor,
              ),
            ),

            const SizedBox(height: 20),

            // Muestra el mensaje de error si existe (Text)
            if (_error.isNotEmpty)
              Text(
                _error,
                style: const TextStyle(
                  color: Colors.red,
                  fontWeight: FontWeight.bold,
                ),
              ),
          ],
        ),
      ),
    );
  }
}
```

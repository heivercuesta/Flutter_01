

#**4. Column (Organizador de Layout Vertical)**

###Este **widget** es un layout que organiza una lista de widgets hijos verticalmente. Es nuestro contenedor principal para poner los campos y el botón uno debajo del otro.

```python
import 'package:flutter/material.dart';

void main() {
  runApp(const ColumnExample());
}

class ColumnExample extends StatelessWidget {
  const ColumnExample({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      home: Scaffold(
        appBar: AppBar(title: const Text('Ejemplo de Column')),
        body: Center(
          // El widget Column organiza verticalmente
          child: Column(
            mainAxisAlignment: MainAxisAlignment.center, // Centra el contenido verticalmente
            crossAxisAlignment: CrossAxisAlignment.center, // Centra el contenido horizontalmente
            children: const <Widget>[
              // Widget 1
              Text('Primer Elemento', style: TextStyle(fontSize: 20)),
              SizedBox(height: 10), // Un espacio entre elementos
              // Widget 2
              Icon(Icons.star, size: 50, color: Colors.amber),
              SizedBox(height: 10),
              // Widget 3
              Padding(
                padding: EdgeInsets.all(8.0),
                child: LinearProgressIndicator(),
              ),
            ],
          ),
        ),
      ),
    );
  }
}
```

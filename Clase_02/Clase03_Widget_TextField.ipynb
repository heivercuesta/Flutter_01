##Este **widget** permite al usuario ingresar texto. En nuestro caso, lo usaremos para ingresar los n√∫meros.

##**Ejemplo:**



```python

import 'package:flutter/material.dart';

// Controlador para capturar el texto ingresado
final TextEditingController _nameController = TextEditingController();

void main() {
  runApp(const TextFieldExample());
}

class TextFieldExample extends StatelessWidget {
  const TextFieldExample({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      home: Scaffold(
        appBar: AppBar(title: const Text('Ejemplo de TextField')),
        body: Padding(
          padding: const EdgeInsets.all(20.0),
          child: TextField(
            controller: _nameController,
            // Importante para la suma: asegura que aparezca el teclado num√©rico
            keyboardType: TextInputType.number,
            decoration: const InputDecoration(
              border: OutlineInputBorder(),
              labelText: 'Ingresa un n√∫mero aqu√≠',
              hintText: 'Ej: 42',
            ),
          ),
        ),
      ),
    );
  }
}
```

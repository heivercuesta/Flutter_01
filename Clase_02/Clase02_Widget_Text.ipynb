
#**1. Text (Mostrar Texto)**

Este widget es el más básico. Se usa para mostrar una cadena de texto inmutable en la pantalla.


# Propiedades:

image.png


# Codificación:

## Ejemplo




```python
import 'package:flutter/material.dart';

void main() {
  runApp(const TextExample());
}

class TextExample extends StatelessWidget {
  const TextExample({super.key});

  @override
  Widget build(BuildContext context) {
    return const MaterialApp(
      home: Scaffold(
        appBar: AppBar(title: Text('Ejemplo de Text')),
        body: Center(
          // ¡Aquí está el widget Text!
          child: Text(
            'Hola, Mundo de Flutter!',
            style: TextStyle(
              fontSize: 30,
              fontWeight: FontWeight.bold,
              color: Colors.purple,
            ),
          ),
        ),
      ),
    );
  }
}
```



#**3. ElevatedButton (Botón Interactivo)**


###Este **widget** representa un botón elevado que responde a la interacción del usuario. Se usa para desencadenar la acción de suma.

```python
import 'package:flutter/material.dart';

void main() {
  runApp(const ButtonExample());
}

class ButtonExample extends StatelessWidget {
  const ButtonExample({super.key});

  // Función que se ejecuta cuando se presiona el botón
  void _handleClick() {
    // Usaremos print en el ejemplo sencillo
    print('¡Botón Sumar Presionado!');
  }

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      home: Scaffold(
        appBar: AppBar(title: const Text('Ejemplo de ElevatedButton')),
        body: Center(
          child: ElevatedButton(
            // Asignamos la función de callback a onPressed
            onPressed: _handleClick,
            child: const Text(
              'Hacer Clic Aquí',
              style: TextStyle(fontSize: 18),
            ),
          ),
        ),
      ),
    );
  }
}
```
